```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
bb = pd.read_csv('./data/billboard.csv')
bb.head(2)
```

```python
# First unpivot week columns into week, rating (cols => rows)
bb_melt = bb.melt(
            id_vars=['year', 'artist', 'track', 'time', 'date.entered']
            , value_vars=None
            , var_name='week'
            , value_name='rating'
        )
bb_melt[bb_melt.track == 'Loser'].sample(5)
```

# 3rd Normal Form

```python
# Subset out songs
songs = bb_melt.loc[:, ['year', 'artist', 'track', 'time']]
songs[songs.track == 'Loser'].sample(5)
```

```python
# Take distinct
songs = songs.drop_duplicates()
songs[songs.track == 'Loser']
```

```python
# Assign primary key
songs['id'] = songs.index # or 
#songs['id2'] = range(songs.shape[0])
songs.head(5)
```

# Joining

```python
# Inner join to bring in song id
ratings = bb_melt.merge(
            right=songs
            , how='inner'
            , on=['year', 'artist', 'track', 'time']
        )

ratings = ratings.loc[:, ['id', 'time', 'date.entered', 'week', 'rating']]
ratings.sample(5)
```
